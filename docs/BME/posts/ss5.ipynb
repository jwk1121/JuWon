# Fourier Transform (1)

## Fourier Transform의 필요성  
* 신호를 주파수 도메인에서 해석하기 위해서는 푸리에 변화에 대해 잘 알고 있어야 함.  
* 푸리에 변환은 푸리에 급수의 확장이다. 
  * 주기성이 있는 신호라면 푸리에 급수, 푸리에 변환이 모두 가능하지만, 주기성이 없는 신호라면 푸리에 변환만을 통해 주파수 도메인에서 해석이 가능하다.  

## Idea for Fourier Transform  
Fourier Transform을 간단히 말하면 주기($T_0$)가 $\infty$인 신호의 Fourier coefficient를 구하는 과정이다. 이를 수식으로 표현하면 아래와 같다.  

$$x(t) = \sum^{\infty}_{k=-\infty}C_ke^{jk\omega_0t},\quad C_k = \frac{1}{T_0}\int_{T_0}x(t)e^{-jk\omega_0t}dt$$

$$C_{k\infty} = \lim_{T_0 \to \infty}\frac{1}{2\pi}\frac{2\pi}{T_0}\int_{T_0}x(t)e^{-jk\omega_0t}dt$$  

주기 ($T_0$)가 무한대에 가까워 지는 것이므로, 
$\displaystyle\lim_{T_0 \to \infty}\frac{2\pi}{T_0} = \lim_{T_0 \to \infty}\omega_0 = d\omega$이고, $\;$ $k\omega_0 \to kd\omega \to \omega$이 된다.

$$ C_{k\infty} = \frac{1}{2\pi}\left[\int_{-\infty}^{\infty}x(t)e^{-j\omega t}dt\right]d\omega$$

정리하면 푸리에 변환과 역 푸리에 변환 식은 아래와 같이 정의할 수 있다.  
$$\;\mathcal{F}[x(t)] = X(\omega) = \int_{-\infty}^{\infty}x(t)e^{-j\omega t}dt$$  

$$\mathcal{F}^{-1}[X(\omega)] = x(t) = \frac{1}{2\pi}\int^{\infty}_{-\infty}F(\omega)e^{j\omega t}d\omega$$   

## Example of Fourier Transform  
푸리에 변환을 하다보면 자주 접하는 함수들의 예시를 다뤄보도록 하겠다. 이번에 다루는 함수들의 푸리에 변환을 암기해두면 복잡한 식들도 쉽게 풀 수 있다.  

### rectangular pulse
주기가 T이고 크기가 V인 사각파는 $f(t) = V rect\left(\frac{t}{T}\right)$ 이러한 수식으로 표현할 수 있다.  이를 푸리에 변환하면 식은 아래와 같다.  
$$F(\omega) = \int_{-\infty}^{\infty}f(t)e^{-j\omega t} dt = \int_{-T/2}^{T/2}Ve^{-j\omega t} dt = \left[\frac{V}{-j\omega}e^{-j\omega t}\right]^{T/2}_{-T/2}= \cdots = TVsinc\left(\frac{T\omega}{2}\right)$$
$$\;\therefore rect\left(\frac{t}{T}\right) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; Tsinc\left(\frac{T\omega}{2}\right) $$  

### Impulse function
$f(t) = A\delta(t - t_0)$를 푸리에 변환하면 식은 아래와 같다.  
$$F(\omega) = \int_{-\infty}^{\infty}f(t)e^{-j\omega t} dt = A\int_{-\infty}^{\infty}\delta(t - t_0)e^{-j\omega t} dt = A e^{-j\omega t_0}$$
$$\;\therefore \delta(t) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; 1$$  

### Inverse Fourier Transform of Impulse function  
주파수 도메인 상의 임펄스 함수 $F(\omega) = 2\pi\delta(\omega - \omega_0)$을 역 푸리에 변환하면 아래와 같은 결과가 나온다.  
$$f(t) = \frac{1}{2\pi}\int_{-\infty}^{\infty}F(\omega)e^{j\omega t} d\omega = \frac{1}{2\pi}\int_{-\infty}^{\infty}2\pi\delta(\omega - \omega_0)e^{j\omega t} d\omega = e^{j\omega_0 t}$$

$$\;\therefore e^{j\omega_0 t}\;\xleftrightarrow{\quad\mathcal{F}\quad}\; 2\pi\delta(\omega - \omega_0)$$


## Properties of the Fourier Transform  

### Linearity
$$[af_1(t) + bf_2(t)] \;\xleftrightarrow{\quad\mathcal{F}\quad}\; [aF_1(\omega) + bF_2(\omega)]$$

### Time scaling
$$f(at) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; \frac{1}{|a|}F\left(\frac{\omega}{a}\right)$$

### Time shifting  
$$f(t-t_0) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; F(\omega)e^{-j\omega t_0}$$

### Time revarsal  
$$f(-t) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; \frac{1}{|-1|}F\left(\frac{\omega}{-1}\right) = F(-\omega)$$

### Time Transformation (= scailing + shifting)
$$f(at-t_0) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; \frac{1}{|a|}F\left(\frac{\omega}{a}\right)e^{-jt_0(\omega/a)}$$

### Duality (쌍대성)
$$F(t) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; 2\pi f(-\omega)\quad when \;\; f(t)  \;
\xleftrightarrow{\;\mathcal{F}\;}\;  F(\omega)$$

### Convolution  
시간 영역에서의 컨볼루션 연산은 주파수 영역에서는 곱셈을 의미한다.  
$$f_1(t) * f_2(t) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; F_1(\omega)F_2(\omega)$$  

반대로 시간영역에서의 곱셈 연산은 주파수 영역에서의 컨볼루션 연산을 의미한다. 단,  
$$f_1(t) f_2(t) \;\xleftrightarrow{\quad\mathcal{F}\quad}\; \frac{1}{2\pi} F_1(\omega)*F_2(\omega)$$  

### Frequency shifting  
$$f(t)e^{j\omega t_0} \;\xleftrightarrow{\quad\mathcal{F}\quad}\; F(\omega - \omega_0)$$

### Time integration  
$$g(t) = \int_{-\infty}^{t}f(\tau)d\tau \;\xleftrightarrow{\quad\mathcal{F}\quad}\; \frac{F(\omega)}{j\omega} + \pi F(0)\delta(\omega) = G(\omega)$$

### Time differentiation  
$$\frac{d^n[f(t)]}{dt}\;\xleftrightarrow{\quad\mathcal{F}\quad}\;(j\omega)^nF(\omega)$$  
$$\because \; f(t) = \frac{1}{2\pi}\int F(\omega)e^{j\omega t}d\omega,\; f^\prime(t) = \frac{1}{2\pi}\frac{1}{j\omega}\int F(\omega)e^{j\omega t}d\omega$$
