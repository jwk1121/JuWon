## 주기 함수(periodic signals)  
주기(period)가 T인 신호는 아래와 같은 수식을 만족한다.  
$$x(t) = x(t + T), T > 0$$  
가장 작은 주기 T를 기본 주기(fundamental period)라 부르고 $T_0$으로 표현한다. 기본 주기가 $T_0$라면 아래와 같은 식을 통해 기본 주파수를 얻을 수 있다.   

$$f_0 = \frac{1}{T_0}Hz$$  
$$w = 2\pi f_0 = \frac{2\pi}{T_0}rad/s$$

### 주기 신호의 덧셈  
$$x(t) = x_1(t)+x_2(t)+\cdots+x_N(t)$$  
위 식에서 $x(t)$의 주기를 알기 위해서는 아래와 같은 과정을 통해 알 수 있다.  
*(각 신호의 주기 비율이 유리수(정수)배 일 때는 주기신호 끼리 더하면 x(t)는 주기 신호임을 알 수 있음.)*  
> 1. 각 신호들의 주기를 파악  
> 2. $T_1/T_i$를 기약 분수 꼴로 구하기 $\therefore i-1$개의 기약분수를 얻게 됨  
> 3. 2과정을 통해 얻은 기약 분수들의 최소공배수 $k_0$를 찾기
> 4. $x(t)$의 주기 $T_0$ 다음과 같음.  
> $$T_0 = k_0 \cdot T_{01}$$

## 공학에서의 exponential fuction이란  
$x(t) = Ce^{at}$가 있을 때 C와 a의 형태에 따른 $x(t)$의 변화에 대해 다룰 예정이다.  

### C와 a가 실수(real)일때  
> i. a > 0일때, 지수함수는 증가하는 형태  
> ii. a = 0일때, t와 무관하게 계속 $x(t)= C$  
> iii. a < 0일때, 지수함수는 감소하는 형태

여기서 a < 0일 때의 $x(t)$는 아래와 같다.  
$$x(t) = Ce^{at} = C e^{-t/\tau}, \tau > 0$$  
*$\tau$는 시상수(time constant)를 의미*

위 그림을 통해 알 수 있듯이 $t = 0$ 일 때의 순간 기울기는 $-\frac{C}{\tau}$ 이다. 만약 이러한 기울기로 신호가 감소한다면 $t = \tau$ 일 때 신호는 사라지게 된다.  
$\to Ce^{-1} = 0.368C$ 즉, t가 $\tau$인 시점에서 신호의 진폭은 약 36.8%로 감소하게 된다.  

### C는 복소수, a는 허수일 때  
$$x(t) = Ce^{at}; C = Ae^{j\phi} = A\angle\phi, a=j\omega_0$$  
여기서 $x(t)$는 오일러 공식을 통해 아래와 같이 표현할 수 있다.  
$$x(t) = Ae^{j\phi}e^{j\omega_0 t} = Ae^{j(\omega_0 t+\phi)}$$  
$$= Acos(\omega_0 t+\phi) + jAsin(\omega_0 t+\phi)$$

> 각 주파수가 $k \omega_0$(여기서 k는 정수)인 신호들의 집합을 **harmonics**라 부른다. harmonics는 이후에 나올 푸리에 급수 전개에 사용된다.

### C와 a가 복소수일 때  
$$x(t) = Ce^{at}; C = Ae^{j\phi} = A\angle\phi, a=\sigma_0+j\omega_0$$  
$$x(t)= Ae^{\sigma_0t}cos(\omega_0 t+\phi) + je^{\sigma_0t}Asin(\omega_0 t+\phi) = Re[x(t)] + Im[x(t)]$$  
위 식에서 실수 부분인 $Re[x(t)]$을 살펴보면 A는 x(t)의 진폭의 크기를 결정하고 $\phi$는 위상. $\omega_0$는 각주파수로 회전속도, $\sigma_0$는 신호가 점점 커질지 작아질지를 결정한다.

## 연속 시간 신호 시스템의 성질  
연속 시간 신호와 시스템의 성질에 대해 다룸  
*뒤로 갈수록 점점 중요도가 높아진다고 생각*  

### memory vs. memoryless  
출력 신호 $y(t_0)$가 입력 신호 $x(t_0)$에만 영향을 받는다면 memoryless system이다.  
* 시스템에 memory가 있다면 해당 시스템은 dynamic system(동적 시스템)이라고 불린다. $y(t) = K \cdot \int_{-\infty}^{t} x(\tau) d\tau$는 dynamic system에 속한다.  

* memoryless system은 static system(정적 시스템)이라 부른다. $y(t) = K \cdot x(t)$은 static system에 속한다.  

### Invertible vs. non-invertible  
Invertible을 쉽게 말하면 일대일 함수이다. 그렇기 때문에 $y(t) = x^2(t)$ system의 경우는 non-invertible system이다.  

### Causal vs. NonCausal  
causal은 뜻 그대로 인과 시스템을 의미한다. output 신호의 $t_0$ 시점의 신호가 input 중 $t(t \leq t_0)$에 속한 신호에만 영향을 받는다면 이는 causal system이다.  
그렇다면 반대로 Noncausal system은 미래의 값을 예측 후 이를 다시 반영하는 시스템들을 의미한다.  

### Stable vs. Unstable  
Stable system은 입력 신호가 안정(bounded)됐다면 출력 신호도 안정되는 시스템을 의미한다. 이를 BIBO Stable System이라고 부른다. 
$$|x(t)| \leq M \to |y(t)| \leq R$$  
*x(t)와 y(t)가 유계 신호임을 위와 같이 표현*  

### Time-invariant vs. Time-varying  
입력 신호가 $t$만큼 time shift 됐다면 출력 신호 또한 $t$만큼 time shift되는 system을 Time-invariant system이라고 한다.
* Time-invariant system의 예시로는 $y(t) = e^{x(t)}$가 있다. 만약 $t \to t-t_0$가 된다면 $y(t-t_o) = e^{x(t-t_0)}$이므로 이는 Time-invariant이다. 

* Time-invarying system의 예시로는 $y(t) = e^{-t}x(t)$가 있다. 만약 $t \to t-t_0$가 된다면 $y(t-t_0) = e^{-(t-t_0)}x(t-t_0) = e^{-t+t_0}x(t-t_0)$가 되므로 이는 time varying이다.

### Linear vs. nonlinear  
Linear system은 Additivity와 Homogeneity를 모두 만족, 즉 중첩 원리(principle of superposition)를 만족하는 시스템을 의미한다.  
* Additivity
$$x_1(t) + x_2(t) \to y_1(t) + y_2(t)$$
* Homogeneity
$$ax_1(t) \to ay_1(t)$$
* principle of superposition
$$a_1x_1(t) + a_2x_2(t) \to a_1y_1(t) + a_2y_2(t)$$

## LTI System의 중요성
Linear property와 Time invariant property를 만족하는 시스템을 LTI system이라고 부른다. LTI system은 공학 분야에서 아래와 같은 이유로 매우 중요하다.  
* 대부분의 물리 시스템이 LTI system으로 정확하게 모델링된다.  
* LTI system 방정식은 풀 수 있다.  
* 많은 LTI system의 디자인 방법 또는 분석 방법을 사용할 수 있다.  
* input signal이 sum of functions이라면 LTI System의 output 또한 sum of responses이다.
